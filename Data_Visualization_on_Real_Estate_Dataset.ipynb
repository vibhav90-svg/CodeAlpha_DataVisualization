```python
# ==========================================
# REAL ESTATE DATA VISUALIZATION
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv("/content/dataset_2.csv")

# Display Basic Information
print("First 5 Records")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nColumns")
print(df.columns)

# ------------------------------------------
# Visualization 1: Property Type Distribution
# ------------------------------------------

plt.figure(figsize=(6,4))
sns.countplot(x='Property_Type', data=df)
plt.title("Property Type Distribution")
plt.show()

# ------------------------------------------
# Visualization 2: Furnishing Distribution
# ------------------------------------------

plt.figure(figsize=(6,4))
sns.countplot(x='Furnishing', data=df)
plt.title("Furnishing Status Distribution")
plt.show()

# ------------------------------------------
# Visualization 3: Location Distribution
# ------------------------------------------

plt.figure(figsize=(10,5))
df['Location'].value_counts().plot(kind='bar')
plt.title("Properties by Location")
plt.xlabel("Location")
plt.ylabel("Count")
plt.show()

# ------------------------------------------
# Visualization 4: Price Distribution
# ------------------------------------------

plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.show()

# ------------------------------------------
# Visualization 5: Rooms vs Price
# ------------------------------------------

plt.figure(figsize=(8,5))
sns.scatterplot(x='Rooms', y='Price', data=df)
plt.title("Rooms vs Price")
plt.show()

# ------------------------------------------
# Visualization 6: Area vs Price
# ------------------------------------------

plt.figure(figsize=(8,5))
sns.scatterplot(x='Area_SqFt', y='Price', data=df)
plt.title("Area vs Price")
plt.show()

# ------------------------------------------
# Visualization 7: Property Type vs Price
# ------------------------------------------

plt.figure(figsize=(8,5))
sns.boxplot(x='Property_Type', y='Price', data=df)
plt.title("Property Type vs Price")
plt.show()

# ------------------------------------------
# Visualization 8: Pool vs Price
# ------------------------------------------

plt.figure(figsize=(8,5))
sns.boxplot(x='Has_Pool', y='Price', data=df)
plt.title("Pool Availability vs Price")
plt.show()

# ------------------------------------------
# Visualization 9: Correlation Heatmap
# ------------------------------------------

numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

print("Data Visualization Completed Successfully")
```
